```python
# Cell 2: Import YOLO and Start Training 

from ultralytics import YOLO

data_config_path = '/kaggle/input/rescuevision-aerial-dataset/data.yaml'

# Load a pre-trained YOLOv8 model
model = YOLO('yolov8n.pt')

# Train the model on our new, superior dataset
results = model.train(
    data=data_config_path,
    epochs=25,
    imgsz=640,
    name='yolov8n_aerial_rescue_v2' # A new name for our definitive model
)
```
